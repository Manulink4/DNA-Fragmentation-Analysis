```python
#Paths and directories
import os
from os import listdir
from pathlib import Path
import glob

#dataset
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu

#ML
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

## models
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
```

```python
def create_dataset():
    path = Path('Data/Breast_Cancer/')

    df_control = pd.concat(
        [pd.read_csv(f, sep='\t', header=None, index_col=0) for f in glob.glob(rf'{path}/control/*.breakage')], axis=1
    ).dropna()

    df_cancer = pd.concat(
            [pd.read_csv(f, sep='\t', header=None, index_col=0) for f in glob.glob(rf'{path}/cancer/*.breakage')], axis=1
    ).dropna()

    df = pd.concat([df_control, df_cancer], axis=1).dropna()

    dataframes = [df, df_control, df_cancer]
    return dataframes
```

```python
def mw_test(row):
    length = len(row)
    index = np.argmax(np.diff(sorted(row.values)))+1

    if 1<index<(length-2):
        return mannwhitneyu(list(row)[0:index], list(row)[(index+1):length]).pvalue < 0.07
    else:
        return False

def get_significant_CPG_index(df):
    dataframe = df
    mw_results = dataframe.apply(lambda row: mw_test(row), axis=1)
    return [idx for idx, x in enumerate(mw_results.values) if x]
```

```python
df, df_control, df_cancer = create_dataset()
df
```

```python
significant_cpg_coords = get_significant_CPG_index(df)
df_significant_cpg = df.iloc[significant_cpg_coords]
```

```python
df_final = df_significant_cpg.T
df_final.index = list(range(0,df_final.shape[0]))

print(df_final.shape)
#Control = 0, Cancer = 1
df_final["Target"] =  [0] * df_control.shape[1] + [1] * df_cancer.shape[1]
df_final

# df_final.to_csv('significant_cpg_python.csv',index=False)
```

```python
X = df_final.loc[:, df_final.columns != 'Target']
y = df_final["Target"]
```

```python
clf = make_pipeline(StandardScaler(), svm.SVC(kernel='linear'))
# svm: svm.SVC(kernel='linear') 0.70


iterations = 250
all_y_true = []
all_y_pred = []

for i in range(iterations):
    X_train, X_test, y_train, y_test = train_test_split(
        X,y, test_size=0.05, stratify=y
    )

    for ground_truth in y_test:
        all_y_true.append(ground_truth)

    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    all_y_pred.append(y_pred)

all_y_pred = [list(p) for p in all_y_pred]
filtered_y_pred = [pred for sample in all_y_pred for pred in sample]

confusion_matrix = metrics.confusion_matrix(all_y_true, filtered_y_pred)
tn, fp, fn, tp = confusion_matrix.ravel()

print(confusion_matrix)
print("Accuracy:",  (tp + tn)/(tp+tn+fp+fn))
```
